```python
def n1():
    print('hello world')
def n2():
    print(10/0)
def n3():
    print(20+30)
```

```python
n1()
n2()
n3()
```

```python
from threading import Thread
p1=Thread(target=n1,name="n1")
p2=Thread(target=n2,name="n2")
p3=Thread(target=n3,name="n3")
```

```python
p1.start()
p2.start()
p3.start()
```
