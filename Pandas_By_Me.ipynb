```python
null=list(db2['Eps'])
print(null)
```

```python
with_nan=[]
for i in null:
    if str(i)=='nan':
        pass
    else:
        with_nan.append(i)
print(with_nan)
        
```

```python
import numpy as np
```

```python
mean=np.sum(with_nan)/len(with_nan)
mean
```

```python
data_2=[]
for i in null:
    if str(i)=='nan':
        data_2.append(mean)
    else:
        data_2.append(i)
print(data_2)
```
