```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
import pandas as pb
```

```python
db1=pb.read_csv("C:\\Users\\Paramjeet\\Downloads\\archive\\tips.csv")
```

```python
db1
```

```python
x=db1.Total_bill
```

```python
y=db1.Tip
```

```python
plt.plot(x,y)
plt.show()
```

```python
salary=[1000,2000,30000,40000,5000]
exp=[1,2,3,4,5]
```

```python
plt.xlabel('salary',c='pink')
plt.ylabel('experenice',c='gold')
plt.xticks([8000,16000,24000,32000,40000])
plt.yticks([1,2,3,4,5])
plt.plot(salary,exp,marker='D',mfc='red',mec='red',markersize=6,c='magenta',linewidth=3.45754545)
plt.title('salary vs experenice',c='orange')
plt.show()
```

```python
plt.title('TOTAL BILL VS TIPS',c='CYAN')
sb.lineplot(x=x,y=y)
plt.show()
```

## BARGRAPH :-

```python
db1.Sex.value_counts()
```

```python
db1.Sex.value_counts().index
```

```python
db1.Sex.value_counts().values
```

```python
plt.bar(x=db1.Sex.value_counts().index,height=db1.Sex.value_counts().values)
plt.show()
```

```python
sb.barplot(x=db1.Sex.value_counts().index,y=db1.Sex.value_counts().values)
plt.show()
```

## Scatter graph :-

```python
plt.scatter(x=db1.Total_bill,y=db1.Tip)
plt.show()
```

```python
sb.scatterplot(x=db1.Total_bill,y=db1.Tip,hue=db1.Sex,style=db1.Time)
plt.show()
plt.figure(figsize=(20,100))
```
